### 10. Ambition box(Salary)

```python
## searching the driver
driver = webdriver.Chrome(r'C:\Users\DELL\OneDrive\Desktop\chromedriver.exe')
url ='https://www.ambitionbox.com/'
driver.get(url)
salary = driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]')
salary
salary.click()
search_job = driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')
search_job
search_job.send_keys('Data Scientist')
search_btn = driver.find_element_by_class_name('icon-search')
search_btn
search_btn.click()
```

```python
## scrapping the data
comp_tags = driver.find_elements_by_xpath('//div[@class="name"]')
comp = []
for i in comp_tags:
    comp.append(i.text)

exp_tags = driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
exp = []
for i in exp_tags:
    exp.append(i.text)
    
avg_tags = driver.find_elements_by_xpath('//p[@class="averageCtc"]')
avg = []
for i in avg_tags:
    avg.append(i.text)

min_tags = driver.find_elements_by_xpath('//div[@class="salary-values"]')
mini = []
for i in min_tags:
    mini.append(i.text)

df = pd.DataFrame()
df['Company']= comp
df['Experience'] = exp
df['Average Sal'] = avg
df['Min/Max Sal']= mini

df =df[:10]
df = df.replace(r'\n','-',regex=True)
df
```
