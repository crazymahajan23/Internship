```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
## Header Tags from Wikipedia.org

page = requests.get('https://en.wikipedia.org/wiki/Wikipedia')

soup = BeautifulSoup(page.content)

title = []
for i in soup.find_all(['h1','h2','h3','h4','h5','h6']):
    title.append(i.text)
title

```

```python
## IMDB top 100 movies
page = requests.get('https://www.imdb.com/list/ls091520106/')
soup = BeautifulSoup(page.content)
Name= []
for i in soup.find_all('h3',class_="lister-item-header"):
    Name.append(i.text)
    
Rating= []
for i in soup.find_all('div',class_="ipl-rating-star small"):
    Rating.append(i.text)
    
Year= []
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    Year.append(i.text)

df= pd.DataFrame({'Name':Name,'Rating':Rating,'Year':Year})

df = df.replace(r'\n','',regex=True)
df
```

```python
## Meesho.com

page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
soup = BeautifulSoup(page.content)

P_Name = []
for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS" ):
    P_Name.append(i.text)

Price =[]
for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    Price.append(i.text)

Discount = []
for i in soup.find_all('span',class_='Text__StyledText-sc-oo0kvp-0 lnonyH'):
    Discount.append(i.text)

df = pd.DataFrame({'P_Name':P_Name,'Price':Price,'Discount':Discount})
df
```

```python
## Coreyms.com

page = requests.get('https://coreyms.com/')
soup = BeautifulSoup(page.content)

Heading = []
for i in soup.find_all('h2',class_='entry-title'):
    Heading.append(i.text)

Date = []
for i in soup.find_all('time',class_='entry-time'):
    Date.append(i.text)

Content = []
for i in soup.find_all('div',class_='entry-content'):
    Content.append(i.text)

Code = []
for i in soup.find_all('iframe',class_='youtube-player'):
    Code.append(i['src'])
Code

df = pd.DataFrame({'Heading':Heading,'Date':Date,'Content':Content})
df = df.drop(index=[4])
df

df.loc[:,'Code']=Code
df.replace(r'\n','',regex=True)
df
```

```python
## Dineout

page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
soup = BeautifulSoup(page.content)

location = []

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location

titles = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
titles

image = []

for i in soup.find_all('img',class_='no-img'):
    image.append(i['data-src'])
image

cuisine = []
for i in soup.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text)
    
rating = []
for i in soup.find_all('div',class_='restnt-rating rating-4'):
    rating.append(i.text)

df = pd.DataFrame({'Title':titles,'Location':location,'cuisine':cuisine,'Image':image,'Rating':rating})
df
```

```python
### bevkoof 

page = requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts&sort=popular')
soup = BeautifulSoup(page.content)

name = []
for i in soup.find_all('div',class_='productCardDetail'):
    name.append(i.text)
    
price = []
for i in soup.find_all('span',class_='discountedPriceText'):
    price.append(i.text)

img = []
for i in soup.find_all('img',class_='productImgTag'):
    img.append(i['src'])
 
df = pd.DataFrame({'name':name,'price':price,'Image':img})
df
```
